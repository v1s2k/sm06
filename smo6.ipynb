```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t_obsl1 = np.linspace(1, 5, num=50)
t_obsl2 = np.linspace(1, 5, num=50)
t_obsl1_grid, t_obsl2_grid = np.meshgrid(t_obsl1, t_obsl2)
lmbd1 = 0.1
lmbd2 = 0.1
mu1 = 1/t_obsl1_grid
mu2 = 1/t_obsl2_grid
y1 = lmbd1 / mu1
y2 = lmbd2 / mu2
y = y1 + y2

tque1 = y1 / (mu1 * (1 - y1))
tsys1 = tque1 + t_obsl1_grid

tque2 = (((mu2 / mu1) * (y1 / (1 - y)) + y) / (1 - y))*(1/mu2)
tsys2 = tque2 + t_obsl2_grid

print (tque2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(t_obsl1_grid, t_obsl2_grid, tsys1, cmap='viridis')
ax.set_xlabel('t_obsl1')
ax.set_ylabel('t_obsl2')
ax.set_zlabel('      I  ')
ax.set_title(' 1')
plt.show()

fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.plot_trisurf(t_obsl1_grid.flatten(), t_obsl2_grid.flatten(), tsys2.flatten(), color='green')
ax.view_init(-140, 30)
ax2.set_xlabel('t_obsl1')
ax2.set_ylabel('t_obsl2')
ax2.set_zlabel('    ')
ax2.set_title('  2')

plt.show()

```
